```python
import numpy as np
import keras
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
%matplotlib inline
```

```python
n_pts=500
np.random.seed(0)

Xa=np.array([np.random.normal(13,2,n_pts),np.random.normal(12,2,n_pts)]).T # top region
Xb=np.array([np.random.normal(8,2,n_pts),np.random.normal(6,2,n_pts)]).T # bottom region

X=np.vstack((Xa,Xb)) # stacking data - top region points and then bottom region points below
Y=np.matrix(np.append(np.zeros(n_pts),np.ones(n_pts))).T # Labels/ground truth - top is 0 and bottom is 1
```

```python
plt.scatter(X[:n_pts,0],X[:n_pts,1]) # top region points (x,y)
plt.scatter(X[n_pts:,0],X[n_pts:,1]) # bottom region points (x,y)
plt.show()
```

```python
model = Sequential()
model.add(Dense(units=1,input_shape=(2,),activation='sigmoid')) # creating a model
adam=Adam(learning_rate=0.1) # optimizer initialization
model.compile(adam, loss='binary_crossentropy', metrics=['accuracy'])
h=model.fit(X,Y,verbose=1,batch_size=50,epochs=500,shuffle='true') #fitting the data into the model for training
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(h.history['loss'])
plt.legend(['loss'])
plt.title('loss')

plt.subplot(1,2,2)
plt.plot(h.history['accuracy'])
plt.xlabel('epoch')
plt.title('accuracy')
plt.ylabel('accuracy (%)')


plt.show()
```

```python
def decision_boundary(x,y,model):
    x_span=np.linspace(min(x[:,0])-1, max(x[:,0])+1) # least and highest point in x-axis with tolerance of +-1
    y_span=np.linspace(min(x[:,1])-1, max(x[:,1])+1) # least and highest point in y-axis with tolerance of +-1
    xx,yy=np.meshgrid(x_span,y_span) # creating a grid of nXn from linspace data
    xx_,yy_=xx.ravel(),yy.ravel() # creating 1D array from 2D meshgrid
    grid=np.c_[xx_,yy_]
    pred_func=model.predict(grid)
    z=pred_func.reshape(xx.shape)
    plt.contourf(xx,yy,z)
```

```python
decision_boundary(X,Y,model)
```

```python
decision_boundary(X,Y,model)
plt.scatter(X[:n_pts,0],X[:n_pts,1]) 
plt.scatter(X[n_pts:,0],X[n_pts:,1])

# predicting a sample point
x=7.5
y=5
point=np.array([[x,y]])
prediction=model.predict(point)
plt.plot(x,y,marker='o',markersize=10,color='r')
print('Prediction is: ',prediction)
```
